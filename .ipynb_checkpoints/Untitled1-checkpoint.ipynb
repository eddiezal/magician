```python
import os
import pandas as pd
import numpy as np
from google.cloud import bigquery, bigquery_storage
import matplotlib.pyplot as plt
import seaborn as sns

# Set up authentication for BigQuery
CREDENTIALS_PATH = r"C:\Users\eddie\OneDrive\code\magician\config\cloud_credentials.json"
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = CREDENTIALS_PATH

# Initialize BigQuery Client
client = bigquery.Client()
bqstorage_client = bigquery_storage.BigQueryReadClient()
```

```python
PROJECT_ID = "cloud4marketing-281206"
DATASET_ID = "crypto_price"
TECHNICALS_TABLE = f"{PROJECT_ID}.{DATASET_ID}.technical_indicators"

query = f"""
    SELECT * FROM `{TECHNICALS_TABLE}`
    WHERE timestamp >= TIMESTAMP_SECONDS(UNIX_SECONDS(CURRENT_TIMESTAMP()) - 6 * 30 * 24 * 60 * 60)
"""

df_indicators = client.query(query).to_dataframe()
print("✅ Data Loaded!")

```

```python
# Check for missing values
print(df_indicators.isnull().sum())

# Check the distribution of indicators
df_indicators.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Choose key indicators to visualize
indicators_to_check = [
    "rsi_14", "macd", "bollinger_upper", "bollinger_lower", 
    "ema_20", "ema_50", "mfi_14", "z_score"
]


# Create subplots
fig, axes = plt.subplots(len(indicators_to_check), 1, figsize=(12, 8))

# Plot histograms
for i, indicator in enumerate(indicators_to_check):
    sns.histplot(df_indicators[indicator].dropna(), bins=50, kde=True, ax=axes[i])
    axes[i].set_title(indicator)

plt.tight_layout()
plt.show()
```
